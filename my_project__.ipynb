 ## **16) VISUALIZING DATA USING BAR CHART**

```python
Productline.plot(kind='bar', rot=1, color=['r','g','b',], figsize=(10,5))
plt.title("Quantity of Productline\n", size=15)
plt.ylabel("Counts")
```

## **17) VISUALIZING DATA USING PIE CHART**

```python
plt.pie(Productline,
        labels=['Motorcycles', 'Classic Cars', 'Trucks and Buses', 'Vintage Cars',
       'Planes', 'Ships', 'Trains'],
        autopct="%0.1f%%",
        explode=[0,.01,0.,0,0,0,0,],
        startangle=60,radius=1)
plt.title("Quantity of Productline", size=15)
plt.show()
```
